# World Plot for Happiness Score for 2017 

```python

import plotly.graph_objs as go 
from plotly.offline import init_notebook_mode,iplot,plot
init_notebook_mode(connected=True) 

table_17.head(10)

table_to_map= table_17[['Country', 'Happiness.Score']].copy()
```

```python
data = dict(
        type = 'choropleth',
        colorscale = 'Viridis',
        reversescale = True,
        locations = table_to_map['Country'],
        locationmode = "country names",
        z = table_to_map['Happiness.Score'],
        text = table_to_map['Country'],
        colorbar = {'title' : 'Happiness Score in 2017'},
      ) 

layout = dict(title = 'Happiness Score',
                geo = dict(showframe = False,projection = {'type':'mercator'})
             )
```

```python
choromap = go.Figure(data = [data],layout = layout)
plot(choromap,validate=False)
```
